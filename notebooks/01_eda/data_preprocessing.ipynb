```python
import pandas as pd
import json
from pathlib import Path

BASE_DIR = Path.cwd().parents[1]  # notebooks/01_eda/ -> up 2 = project root
json_path = BASE_DIR / "data" / "raw" / "dbaasp_full.json"

print("Loading from:", json_path)

with open(json_path) as f:
    data = json.load(f)
```

```python
import json
from IPython.display import display, JSON

display(JSON(data[:3], expanded=True))
```

```python
df = pd.json_normalize(
    data, 
    record_path=['targetActivities'], 
    meta=['sequence']
)
```

```python
columns_to_extract = [
    'sequence', 
    'targetSpecies.name', 
    'activityMeasureGroup.name',
    'unit.name',
    'activity'
]
```

```python
extracted_df = df[columns_to_extract]
```

```python
extracted_df.columns = [
    'sequence', 
    'target_activity_name', 
    'activity_measure_group', 
    'unit',
    'activity'
]
```

```python
# 4. Export to CSV
extracted_df.to_csv('../../data/raw/extracted_peptides.csv', index=False)

# Display the output in Notebook
display(extracted_df)
```
